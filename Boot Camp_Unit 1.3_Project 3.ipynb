<b>1. Choose one variable and plot that variable four different ways.</b>

```python
# Importing data.
min_wage = pd.read_csv('Minimum Wage Data.csv',encoding = 'Windows-1252')

mean_yearly_wage = min_wage.groupby(['Year'], as_index=False).mean()
mean_yearly_wage.head()
```

```python
f, ax = plt.subplots(figsize=(30,10))
sns.set(style = 'darkgrid')

# Plotting a line.
plt.subplot(141)
sns.lineplot(range(len(mean_yearly_wage)),mean_yearly_wage['High_Value'])
plt.title('Plot 1:Line Plot')

# Plotting scatter plot.
plt.subplot(142)
sns.boxplot(mean_yearly_wage['High_Value']) 
plt.title('Plot 2: Box Plot')

# Plotting bar plot.
plt.subplot(143)
sns.distplot(mean_yearly_wage['High_Value'])
plt.title('Plot 3: Histogram')

# Plotting violin plot.
plt.subplot(144)
sns.violinplot(mean_yearly_wage['High_Value'])
plt.title('Plot 4: Violin Plot')

plt.show()
```

<font style = 'font-family:calibri'><font size = 5>Using a <b>line plot</b> without a categorical value to plot a continous variable may not be the best method to display data. A line plot is primarily used to display change over time. From this chart we can see the minimum and maximum values. However, we are not able to determine over what time these changes occured.
<p> A <b>box plot</b> does a better job at representing one continuous variable. In this form we can see the min, max and mean of the variable. We can also see the distribition of the variable and if there are any outliers.
<p><b>Histograms</b> are another great way to view the distribution of one variable. This also lets us see the central tendency, variability as well as outliers.
<p>A <b>voilin plot</b> is another good way of displaying the shape of one variable. We can use a violing plot to show the probability density.</p>

<b>2. Choose two continuous variables, and plot them three different ways.</b>

```python
figsize=(20,10)

g = sns.lmplot(y = 'High_Value',
               x = 'High_2018',
               data = mean_yearly_wage,
               fit_reg = False, 
               scatter_kws = {'alpha':0.4})
g.set_ylabels('High_Value')
g.set_xlabels('High_2018')
plt.title('Plot 5: Scatterplot,Yearly Wage Compared to 2018 Value')
plt.show()

g2 = sns.jointplot(y = 'High_Value',
                  x = 'High_2018',
                  kind = 'kde',
                  data = mean_yearly_wage)
plt.subplots_adjust(top=1.4)
plt.title('Plot 6: Joint Plot,Yearly Wage Compared to 2018 Value')
plt.show()

g3 = sns.distplot(mean_yearly_wage['High_Value'])
sns.distplot(mean_yearly_wage['High_2018'])
plt.title('Plot 7: Histogram')
plt.show()
```

<font style = 'font-family:calibri'><font size = 5><b> Plot 5</b> is useful when comparing values and can be useful to predict future values. The jointplot <b>Plot 6</b> is a good way to see the shape of each value. The kernal density plot in the center can al tell us where the majority of the values sit. <b>Plot 7</b> also displays the shape of the data. If one was looking for more detailed information, <b>Plot's 6</b> and <b>7</b> would not be the best use.

<b>3. Choose one continuous variable and one categorical variable, and plot them six different ways.</b>

```python
figure, ax = plt.subplots(figsize=(40,10))
sns.set(style = 'darkgrid')


# Plotting a box plot for High_Value and Year.
plt.subplot(271)
sns.boxplot(round(mean_yearly_wage['High_Value']), 
            mean_yearly_wage['Year'])
plt.xticks(rotation=80)
plt.title('Plot 8: Box Plot')

# Plotting scatter plot of High_Value and Year.
plt.subplot(272)
sns.scatterplot(x = 'Year',
                y = 'High_Value',
                data = mean_yearly_wage)
plt.title('Plot 9: Scatter Plot')

# Plotting a regression line.
plt.subplot(273)
sns.regplot(x = 'Year',
            y = 'High_Value',
            data = mean_yearly_wage)
plt.title('Plot 10: Regression Plot')


figure, ax = plt.subplots(figsize=(40,10))

# Plotting joint plot with regression line.
plt.subplot(271)
sns.lineplot(x ='Year',
             y = 'High_Value',
             data = mean_yearly_wage)
plt.title('Plot 11: Joint Plot')

plt.subplot(272)
sns.barplot (x ='Year',
             y = 'High_Value',
             data = mean_yearly_wage)
plt.xticks(rotation=85)
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=5))
plt.title('Plot 12: Bar Graph')

plt.subplot(273)
sns.pointplot(x = 'Year',
              y = 'High_Value',
            kind = 'point',
            data = mean_yearly_wage)
plt.xticks(rotation=85)
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=5))
plt.title('Plot 13: Point Plot')
plt.show()
```

<font style = 'font-family:calibri'><font size = 5> <b>Plots, 9, 10, 11</b> and <b> 13 </b>all give similar information. These graphs give us the the trend over time. These might be the best graphs to use when displaying one continious variable an one categorical value. <b>Plot 12: Bar Graph</b> can also be another great way to display this kind of data. Because of the large number of dates, from 1968 to 2017, it may not be as visually readible as the various line plots.
    <p><b>Plot 8</b> displays the disptribution in wages for each year. This graph, thou potentialy useful, is also not very readible. There are alot of dates. It might be better if only one year was displayed at a time.
